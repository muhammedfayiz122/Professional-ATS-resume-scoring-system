```python
from langchain.document_loaders import unstructured
loader = unstructured.PDFLoader(path)
docs = loader.load()
docs[0].page_content
```

```python
from langchain_unstructured import UnstructuredLoader

loader = UnstructuredLoader(
    file_path=path,
    strategy="hi_res",
    partition_via_api=True,
    coordinates=True,
    
)
docs = []
for doc in loader.lazy_load():
    docs.append(doc)
docs
```

```python
import fitz  # PyMuPDF

doc = fitz.open(path)
all_text = []

for page in doc:
    blocks = page.get_text("dict")["blocks"] #type: ignore
    for block in blocks:
        for line in block.get("lines", []):
            for span in line["spans"]:
                text = span["text"].strip()
                if text:
                    all_text.append(text)
```

```python
resume_text = "\n".join(all_text)
print(resume_text)
```

```python
from fitz import pymupdf
doc = pymupdf.open(path)
header = "Header"  # text in header
footer = "Page %i of %i"  # text in footer
for page in doc:
    page.insert_text((50, 50), header)  # insert header
    page.insert_text(  # insert footer 50 points above page bottom
        (50, page.rect.height - 50),
        footer % (page.number + 1, doc.page_count),
    )
    print(page.get_text())
```
